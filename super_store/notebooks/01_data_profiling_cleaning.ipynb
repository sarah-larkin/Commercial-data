## Validate and Clean the data 

### Datatype conversion

```python
#converting date str to datetime
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True)
df['ship_date'] = pd.to_datetime(df['ship_date'], dayfirst=True)
df.head()
```

```python
df[['order_date', 'ship_date']].dtypes
```

```python
#removing comma's for numbers in thousands and converting sales column type from str to int 
df['sales'] = df['sales'].replace({',':''}, regex=True).astype(float)
df['sales'].dtype
```

### Checking each states relatetes to one region only 

```python
#check each state only relates to one region

states_df = df_locations.groupby('state')['region']\
    .nunique()\
    .reset_index(name='number_of_regions')\
    .sort_values('number_of_regions', ascending=False)

#view any states that map to more than 1 region
states_df[states_df['number_of_regions'] > 1]

#Initially converted to csv then added comments manually 
```

### Detailing all states that map to more than one region in one combined table 

```python
#count of rows per location combination (similar to above)
state_region_market_counts = (
    df_locations
        .groupby(['state', 'region', 'market'])
        .size()
        .reset_index(name='No_of_rows')
)

#count how many regions per state to check each state only relates to one region (similar to above)
region_counts = (
    state_region_market_counts
        .groupby('state')['region']
        .nunique()
        .reset_index(name='number_of_regions')
)

# join the above together so it can be viewed together
final_df = (
    state_region_market_counts
        .merge(region_counts, on='state')
        .sort_values(['number_of_regions','state', 'region', 'market'], ascending=False)
)

#filter to just the states with more than one region 
final_df = final_df[final_df['number_of_regions'] > 1]

final_df

#export to csv
final_df.to_csv("01_data_checks/01_state_check_2.csv", index=False)  #already created a file and updated manually, created this file separately for reference
```

### Checking each region maps to just one market 

```python
# check each regions only relates to one market
regions_df = df.groupby('region')['market'].nunique().reset_index(name='number_of_markets').sort_values('number_of_markets', ascending=False)

regions_df = regions_df[regions_df['number_of_markets'] > 1]

regions_df

```

- All regions that map to multiple markets have generic names (Central, South, North) which is acceptle
- Noted that region should not be viewed in isolation but in relation to the market

```python
#count of rows per location combination (same)
state_region_market_counts = (
    df_locations
        .groupby(['state', 'region', 'market'])
        .size()
        .reset_index(name='No_of_rows')
)

#count how many markets per region to check each region only relates to one market
market_counts = (
    state_region_market_counts
        .groupby('region')['market']
        .nunique()
        .reset_index(name='number_of_markets')
)

# join the above together so it can be viewed together (same)
final_df = (
    state_region_market_counts
        .merge(market_counts, on='region')
        .sort_values(['number_of_markets', 'region', 'market'], ascending=False)
)

#filter to just the regionss with more than one market 
final_df = final_df[final_df['number_of_markets'] > 1]

final_df

#export to csv
final_df.to_csv("01_data_checks/01_region_check.csv", index=False)
```

Description: 
- Markets is the highest geographical level, broken down to region, then state (both nested)
<br><br>

Observations:
- Not all states are unique either due to generic terminology (eg. Central, Southern) or states in different regions having the same name (eg. San Juan in Argentina or Puerto Rico). Therefore states cannot be considered in isolation, but in relation to the region/market.
- It appears that not all states are inputted in the same language (eg. Central, Centrale, Centre or Ouest, Sud). This has shown up where there are duplicates, so indication that locations being listed in the same language may not be reliable throughout the data. 
- Regions are not unique (Generic terms eg.North, South) so should not be considered in isolation but in relation to the market/state
- The markets overlap in terms of geographic location, this is particularly ambiguous EMEA, where EU and Africa are also listed as individual markets. Reason for these groupings is unknown, it is possilbe that there is a business reason behind this eg. some product ranges being for particular regions only (EU/Africa) whereas others have a broader target audience (EMEA)
<br><br>

Assumptions/questions: 
- Assumed that these groupings are defined and understood by the business so not being reclassified at this stage.
- Would be helpful to understand the background of these location groupings from the business particularly where Markets overlap. 
<br><br>

Suggestions: 
- Use a logical location grouping eg. country/continent  
- Ensure location input is in the same language or agreed terminology used for specific locations 
- Possible to clean location further based on agreed groupings (suggestions may become clearer on further analysis) eg. removal of EMEA
